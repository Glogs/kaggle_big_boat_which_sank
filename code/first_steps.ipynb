```python
import pandas as pd
from sklearn.cross_validation import KFold as KFoldOld
from sklearn.model_selection import KFold
```

```python
kf = KFoldOld(100, n_folds=3, random_state=1)
print(len(kf))
```

```python
for a in kf:
    print(">>"+str(a))
```

```python
knf = KFold(n_splits=3, random_state=1)
```

```python
knf.get_n_splits(100)
```

```python
ran = range(13,100)
for a,b in knf.split(ran):
    print(">>"+str(a))
    print("++>>"+str(b))
```

We can see the difference between the deprecated (`cross_validation`) and the new appraoch (`model_selection`) is that the latter is reusable and more of a k-folding strategy, as it gets applied to whichever iterable is passed as argument for `split`.

# Let's play a bit with the data

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
len(train[train['Survived'] != 1])
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"]
```

```python
train
```

```python
import operator

# A dictionary mapping family name to id
family_id_mapping = {}

# A function to get the id given a row
def get_family_id(row):
    # Find the last name by splitting on a comma
    last_name = row["Name"].split(",")[0]
    # Create the family id
    family_id = "{0}{1}".format(last_name, row["FamilySize"])
    print(family_id)
    # Look up the id in the mapping
    if family_id not in family_id_mapping:
        if len(family_id_mapping) == 0:
            current_id = 1
        else:
            # Get the maximum id from the mapping and add one to it if we don't have an id
            current_id = (max(family_id_mapping.items(), key=operator.itemgetter(1))[1] + 1)
        family_id_mapping[family_id] = current_id
    return family_id_mapping[family_id]

# Get the family ids with the apply method
family_ids = train.apply(get_family_id, axis=1)

# There are a lot of family ids, so we'll compress all of the families under 3 members into one code.
#family_ids[train["FamilySize"] < 3] = -1

# Print the count of each unique id.
#print(pd.value_counts(family_ids))

train["FamilyId"] = family_ids
print(train[['FamilyId','FamilySize', 'Name', 'SibSp','Parch']][train['FamilyId'] == 14])

print(train[['FamilyId','FamilySize', 'Name', 'SibSp','Parch']][train['FamilyId'] == 283])

print(train[['FamilyId','FamilySize', 'Name', 'SibSp','Parch']][train['FamilyId'] == 588])
```
